```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import roc_curve, auc
```

```python
inception_df = pd.read_csv("segmented_inceptionv3_roc.csv")
resnet_df = pd.read_csv("segmented_resnet50v2_roc.csv")
vgg_df = pd.read_csv("segmented_vgg16_roc.csv")
```

```python
inception_auc = auc(inception_df["fpr"], inception_df["tpr"])
resnet_auc = auc(resnet_df["fpr"], resnet_df["tpr"])
vgg_auc = auc(vgg_df["fpr"], vgg_df["tpr"])

plt.figure(figsize = [8, 8])
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(inception_df["fpr"], inception_df["tpr"], label='InceptionV3 (AUC = {:.3f})'.format(inception_auc))
plt.plot(resnet_df["fpr"], resnet_df["tpr"], label='Resnet50V2 (AUC = {:.3f})'.format(resnet_auc))
plt.plot(vgg_df["fpr"], vgg_df["tpr"], label='VGG16 (AUC = {:.3f})'.format(vgg_auc))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend(loc='best')
plt.show()
```
